#### 1.2 Cумма элементов


```python
import random
import numpy
import time
import matplotlib.pyplot
from numpy.lib.index_tricks import diag_indices_from
from numpy.ma.core import indices

number = 10**5 * 17
step  = 100 * 17
possible_values = list(range(1, number + 1, step))
times_list = []
n_list= []

for i in range(1,6):
    n = random.choice(possible_values)
    n_list.append(n)

n_list.sort()

for y in n_list:  
    vector = numpy.random.rand(y)
    sum_1 = 0
    start_time = time.time()
    for i in vector:
        sum_1 += i
    end_time = time.time()
    times_list.append(end_time - start_time)
    
matplotlib.pyplot.plot(n_list, times_list)
matplotlib.pyplot.xlabel("Размер вектора (n)")
matplotlib.pyplot.ylabel("Время выполнения (сек)")
matplotlib.pyplot.title("Зависимость времени выполнения от размера вектора")
matplotlib.pyplot.show()
```

#### 1.6 Поиск минимума простым перебором


```python
import random
import numpy
import time
import matplotlib.pyplot

number = 10**5 * 17
step  = 100 * 17
possible_values = list(range(1, number + 1, step))
times_list = []
n_list= []

for i in range(1,6):
    n = random.choice(possible_values)
    n_list.append(n)

n_list.sort()

for dimension in n_list:
    vector = numpy.random.rand(dimension)
    min_number = vector[0]
    start_time = time.time()
    for element in range(1, len(vector)) :
        if element < min_number:
            min_number = element
    end_time = time.time()
    times_list.append(end_time - start_time)

matplotlib.pyplot.plot(n_list, times_list)
matplotlib.pyplot.xlabel("Размер вектора (n)")
matplotlib.pyplot.ylabel("Время выполнения (сек)")
matplotlib.pyplot.title("Зависимость времени выполнения от размера вектора")
matplotlib.pyplot.show()    
```

### Задание 2
Сгенерируйте случайные матрицы $A$ и $B$ размером $n × n$ с неотрицательными элементами. 
Найдите обычное матричное произведение матриц $A$ и $B$.

```python
import random
import numpy

number = 10 * 17 #взяты значение меньше, чтобы не слишком долго ждать
step  = 17
possible_values = list(range(1, number + 1, step))


n = random.choice(possible_values)
matrix_a = numpy.random.random((n,n))
matrix_b = numpy.random.random((n,n))

rows_A = len(matrix_a)
cols_A = len(matrix_a[0])
rows_B = len(matrix_b)
cols_B = len(matrix_b[0])


matrix_result = [[0 for _ in range(cols_B)] for _ in range(rows_A)]

# Перемножение матриц
for i in range(rows_A):
    for j in range(cols_B):
        for k in range(cols_A):
            matrix_result[i][j] += matrix_a[i][k] * matrix_b[k][j]

print(matrix_result)
```
